```python
import numpy as np
import pandas as pd

csv_data_path = 'https://raw.githubusercontent.com/turingplanet/pandas-intro/main/public-datasets/country.csv'
country_info = pd.read_csv(csv_data_path)
country_info
```

```python
# 获取特定列 Country 的内容
country_info['Country']
```

```python
# 获取第2行到第3行的records
country_info[1:3]
```

```python
# 获取第一行的内容
country_info.iloc[1]
```

```python
# 获取 index 为 0 的 record
country_info.loc[0]
```

```python
# 将数据按照国家名正序排列（从A排到Z）
country_info.sort_index()
# 将数据按照国家名正序排列（从Z排到A）
country_info.sort_index(ascending=False)
```

```python
# 将国家名为Zimbabwe和Yemen的数据删除掉，axis = 0 代表根据行来定位数据
# country_info.drop(['Andorra', 'Algeria'], axis = 0) # KeyError 
# 将Region和Population两列的数据全部删除，axis = 1 代表根据列来定位数据
country_info.drop(['Region', 'Population'], axis = 1)
```

```python
# 返回一个Boolean Series，告知我们哪些国家的净迁移率（Net migration）等于'0'
country_info['Net migration']=='0'
# 获得所有0净迁移率国家的数据
zero_migration_filter = (country_info['Net migration'] == '0')
country_info[zero_migration_filter]
# 只查看0净迁移率国家的Region和Populatino列
country_info.loc[(country_info['Net migration']=='0'), ['Region', 'Population']]

# 查看净迁移率小于100的国家信息
litter_migration_filter = (country_info['Net migration'] <= '100')
country_info.loc[litter_migration_filter,['Country']]
```
